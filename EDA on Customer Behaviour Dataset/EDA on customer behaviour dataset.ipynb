```python
value_counts = df['MONTH'].value_counts()

fig, ax = plt.subplots()
bars = ax.bar(value_counts.index, value_counts)

for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Count of each Month')

plt.show()
```
